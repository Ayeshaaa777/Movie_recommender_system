```python
# ast module is being imported and used to convert a string of list into a list
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[0].cast
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter<3:
          L.append(i['name'])
          counter=counter+1
        else:
              break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew'].iloc[0]
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
           L.append(i['name'])
           break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'].iloc[0]
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# sam worthington and sam mendes in the dataset.. can be confusing as tags can be sam,worthington,sam,mendes ..lets remove that
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])

```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags']=movies['overview']+movies['keywords']+movies['genres']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df= movies[['title','tags']]
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
# converting the list into a string
```

 Bag of words vectorization (text to vectors) technique... stop words are words like are,in,is and they are removed


```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

--> picking up 5000 most common words from tags and counting the number of times the words have occured in each movie
--> .toarray() means that the scipy sparse matrix is converted to a numpy array for better visualization

```python
vectors= cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
      y.append(ps.stem(i))
    return " ".join(y)


```

```python
new_df['tags'].apply(stem)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity=cosine_similarity(vectors)
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```

```python
recommend('Avatar')
```
